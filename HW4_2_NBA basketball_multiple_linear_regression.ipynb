### The following data  are for each player. Reference: The official NBA basketball
#### height = height in feet
#### weight = weight in pounds
#### goals = percent of successful field goals 
#### throws = percent of successful free throws 
#### score = average points scored per game
##### https://github.com/mathawanup/master_dataset/blob/master/basketball_score.csv

```python
## ให้นิสิตทดลองทำนาย score ของผู้เล่น เมื่อกำหนด height=6.5 weight=194 goals=0.503 และ throws=0.833
```

```python
import pandas as pd # data manipulation and analysis
import seaborn as sns # data visualization
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
# https://github.com/mathawanup/master_dataset/blob/master/basketball_score.csv
url='https://raw.githubusercontent.com/mathawanup/master_dataset/master/basketball_score.csv'
df=pd.read_csv(url)
df.head() # 5 rows only
```

```python
X = df[["height","weight","goals","throws"]]
y = df['score']
```

```python
import sklearn
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.predict([[6.5,194,0.503,0.833]])
```

```python
fig, axes = plt.subplots(1, 4)
sns.regplot(x='height', y='score', data=df, scatter_kws={'alpha': 0.4},
           line_kws={'color': 'orange'},ax=axes[0])
sns.regplot(x='weight', y='score', data=df, scatter_kws={'alpha': 0.4},
           line_kws={'color': 'red'},ax=axes[1])
sns.regplot(x='goals', y='score', data=df, scatter_kws={'alpha': 0.4},
           line_kws={'color': 'green'},ax=axes[2])
sns.regplot(x='throws', y='score', data=df, scatter_kws={'alpha': 0.4},
           line_kws={'color': 'orange'},ax=axes[3])
plt.show();
```
